# Black Lives Matter - Twitter Sentiment Analysis

### Abstract
In this report we will discuss a possible approach for a Twitter sentiment analysis. In particular this analysis regards the Black Lives Matter movement. This solution lays the foundation about the Natural Language Processing by adopting some of the most common techniques about text analysis, such as tokenization, stemming, stopwords elimination and n-grams implementation

## Problem overview
The current analysis is a sentiment analysis based on a collection of 80.000 tweets collected from the 2020-05-31 to 2020-06-08 about the Black Lives Matter movement. This movement is a decentralized political and social movement advocating for non-violent civil disobedience in protest against incidents of police brutality and all racially motivated violence against black people [1]. It was born in 2013 but it (sadly) becomes well known all over the word after the murder of George Floyd on 2020-05-25.

These tweets are already scraped and labeled as:
- 1, if they shows a positive sentiment towards the movemenet
- 0, if they shows a negative sentiment twoards the movement

Each tweet is described by 28 features and it is possible to study them by the Tweet Object Documentation [2]. The feature regarding the sentiment of the tweet was added for the special purpose of the analysis. For this analysis we focused just on the label and on the relative text, contained in "full_text". 

This problem is well balanced, in fact we have 80.000 tweets, 40.069 classified as 0 and the remaining 39.931. There was a strong activity during this week, in fact we have found in this sample a number of retweets equal to 981.031.288. As you can see in Figure 1, the daily activity increased until 2020-06-03 where it reached its peak and it strongly decreased until the last day to our disposal. 

We have also analyzed the most frequent words inside both categories and you can see it in Figure 2. In particular they are both "coherent" with what we could expect from them. In fact for the "positive" tweets we can see words as "petitions", "everyone", "toghether", while for post against the movement there are words as "white", "division", "behind" and so on. 

## Proposed approach

### A. Data Preprocessing

Natural Language Processing problems have a canonical preprocessing structure, in fact in order to analyze a collection of document we should consider several consective steps. In particular when we deal with huge document, we can actually split them in shorter pieces in order to better analyze their content. Since this is not the case, we skipped to the second step, that is the tokenization, where we split each content into tokens. These tokens will represent the "core" of our analysis, because some of these tokens will be strictly related to our class. In particular we used the Natural Language Tool Kit tokenizer, that allows to easily convert a sentence like "we are equal" into the following array of features ["we","are","equal"]. 

The aim of any sentiment analysis is to extract the most relevant features for each class, and in this case the most relevant ones are the most frequent ones. In order to count the frequency of each token, we need to represents our documents as bag of words, in particular we define a huge sparse matrix that stores counts of all the words in our collection of tweets. To do so, we used the fairly well known TfIdfVectorizer defined in scikit-learn.
TF states for term frequencies and IDF states for inverse document frequency. It means that it penalize the tokens that are frequent in all the documents (our tweets), and it is defined as :

$TfIdf(t) = freq(t,d) * \log{( \frac m {freq(t,D)})}$

where t is the token, d is the current document, and D is the collection of m documents. So it is the basic idea of the sentiment analysis, but there are several other techniques that help increasing the performance. In particular we adopted the stemming and the n-gram implementation. The stemming comes from linguistic and it extract the root form for each word, for example "consult, consulting, consultant" will be treated as "consult", in order to unify analogous words. Then, the n-grams are combinations of words in order, in particular they are defined 



To be more precise, we actually tried to remove also the "stopwords" that are the most common words present in the English grammar (articles, noun, pronoun and so on) but we noticed worse performance. It is due to the fact that probably the stopwords that we used contained some relevant words for our analysis.

## Preprocessing

```python
df.info()
```

```python
df_cleaned = df[["full_text","class"]] 
df_cleaned.sample(2)
```

```python
y = df_cleaned['class']
X = df_cleaned.drop(columns=['class'])
```

```python
english_stemmer = Stemmer.Stemmer('en')

class StemmedTfidfVectorizer(TfidfVectorizer):
     def build_analyzer(self):
            analyzer = super(TfidfVectorizer, self).build_analyzer()
            return lambda doc: english_stemmer.stemWords(analyzer(doc))
```

## Model selection

```python
def doLogisticRegression(x,y):
    
    model = LogisticRegression(max_iter=10000)
    
    parameters ={
        'C':np.logspace(-3,3,7), 
        'penalty':["l1","l2"], # l1 lasso l2 ridge
        'random_state' : [42]
    } 
    
    gs = GridSearchCV(estimator=model,  
                         param_grid=parameters,
                         scoring='accuracy',
                         cv=5,
                         n_jobs=-1)

    gs.fit(x, y)
    print("\n*** LOGISTIC REGRESSION")
    print(f"Best parameters = {gs.best_params_}")
    print(f"Best score = {gs.best_score_ }")
```

```python
def doMultinomialNB(x,y):
    
    model = MultinomialNB()
    
    parameters =  {
        'alpha': np.linspace(0.5, 1.5, 6),
        'fit_prior': [True, False],
        'random_state' : [42]
    }
    
    gs = GridSearchCV(estimator=model,  
                         param_grid=parameters,
                         scoring='accuracy',
                         cv=5,
                         n_jobs=-1)

    gs.fit(x, y)
    print("\n*** MULTINOMIAL NB")
    print(f"Best parameters = {gs.best_params_}")
    print(f"Best score = {gs.best_score_ }")
```

### Naive solution

```python
tfidf_naive = TfidfVectorizer(tokenizer=word_tokenize, token_pattern=None)
X_naive = tfidf_naive.fit_transform(X.values.ravel())
```

```python
doLogisticRegression(X_naive,y)
doMultinomialNB(X_naive,y)
```

### Stopwords removal and Stemming

```python
tfidf = StemmedTfidfVectorizer(min_df=1, stop_words=sw.words('english'), analyzer='word', ngram_range=(1,3))
X_complete = tfidf.fit_transform(X.values.ravel())
```

```python
doLogisticRegression(X_complete,y)
doMultinomialNB(X_complete,y)
```

### Stemming but no stopwords

```python
tfidf_nosw = StemmedTfidfVectorizer(min_df=1, analyzer='word', ngram_range=(1,3))
X_no_sw = tfidf_nosw.fit_transform(X.values.ravel())
```

```python
doLogisticRegression(X_no_sw,y)
doMultinomialNB(X_no_sw,y)
```
